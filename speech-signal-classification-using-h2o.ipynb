# Cartesian Grid Search

```python
from h2o.grid.grid_search import H2OGridSearch
```

```python
param = { 'alpha': [x * 0.01 for x in range(0,11)]
    }

h20_grid = H2OGridSearch(model = H2OGeneralizedLinearEstimator(family = 'binomial'),
                         hyper_params = param, 
                         search_criteria = {'strategy': "Cartesian"},
                         grid_id = 'glm_grid1')
```

```python
h20_grid.train(x = x,y = y,training_frame= train,validation_frame=validate, lambda_search = True)
```

```python
cartesian_grid_perf = h20_grid.get_grid(sort_by='auc', decreasing=True)
cartesian_grid_perf
```

**Thus according to the cartesian grid search, the best value for alpha = 0.06 as it gives the AUC = 0.89166 which is a slight improvement over our previous model with AUC = 0.89.**

```python
# Grabbing the best model (over validation frame) from Cartesian grid search
best_cart_model = cartesian_grid_perf.models[0]
```

# Random Grid Search

```python
params = {
   'alpha': [x * 0.01 for x in range(0,99)]
}
```
